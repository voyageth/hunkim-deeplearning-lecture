# Learning rate = 10

```python
import tensorflow as tf
import numpy as np
from IPython.display import display

xy = np.loadtxt('data/lab06_train.txt', unpack=True, dtype='float32')
x_data = np.transpose(xy[0:3])
y_data = np.transpose(xy[3:])

X = tf.placeholder('float', [None, 3])
Y = tf.placeholder('float', [None, 3])

W = tf.Variable(tf.zeros([3,3]))

hypothesis = tf.nn.softmax(tf.matmul(X, W))
learning_rate = 10
cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(hypothesis), reduction_indices=1))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)

    for step in range(2001):
        sess.run(optimizer, feed_dict={X:x_data, Y:y_data})
        if step % 20 == 0:
            print(step, "cost:%s, W:%s"%(sess.run(cost, feed_dict={X:x_data, Y:y_data}), sess.run(W)))
    
    a = sess.run(hypothesis, feed_dict={X:[[1,11,7]]})
    print("a : %s, arg_max(a) : %s" % (a, sess.run(tf.arg_max(a, 1))))
    b = sess.run(hypothesis, feed_dict={X:[[1,3,4]]})
    print("b : %s, arg_max(c) : %s" % (b, sess.run(tf.arg_max(b, 1))))
    c = sess.run(hypothesis, feed_dict={X:[[1,1,0]]})
    print("c : %s, arg_max(b) : %s" % (c, sess.run(tf.arg_max(c, 1))))
    all_data = sess.run(hypothesis, feed_dict={X:[[1,11,7], [1,3,4], [1,1,0]]})
    print("all_data : %s, arg_max(all_data) : %s" % (all_data, sess.run(tf.arg_max(all_data, 1))))
```

# Load MNIST datasets

```python
from tensorflow.contrib.learn.python.learn.datasets.mnist import read_data_sets
mnist = read_data_sets('./temp', one_hot=True)
```

# MNIST training

```python
x = tf.placeholder('float', [None, 784])
y = tf.placeholder('float', [None, 10])

W = tf.Variable(tf.zeros([784, 10]))
b = tf.Variable(tf.zeros([10]))

activation = tf.nn.softmax(tf.matmul(x, W) + b)

learning_rate=0.01
cost = tf.reduce_mean(-tf.reduce_sum(y * tf.log(activation), reduction_indices=1))
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

init = tf.global_variables_initializer()

training_epochs = 10
batch_size = 10
display_step = 1

sess = tf.Session()
sess.run(init)

for epoch in range(training_epochs):
    avg_cost = 0.
    total_batch = int(mnist.train.num_examples / batch_size)
    for i in range(total_batch):
        batch_xs, batch_ys = mnist.train.next_batch(batch_size=batch_size)
        sess.run(optimizer, feed_dict={x : batch_xs, y: batch_ys})
        avg_cost += sess.run(cost, feed_dict={x : batch_xs, y: batch_ys}) / total_batch

    if epoch % display_step == 0:
        print('Epoch: %04d' % (epoch+1), 'cost={:.9f}'.format(avg_cost))

print('Optimization Finished!')
```

# MNIST Predict & Show

```python
from random import randint

r = randint(0, mnist.test.num_examples - 1)
print('Label: %s' % sess.run(tf.argmax(mnist.test.labels[r:r+1], 1)))
print('Prediction: %s' % sess.run(tf.argmax(activation, 1), {x:mnist.test.images[r:r+1]}))

import matplotlib.pyplot as plt
plt.imshow(mnist.test.images[r:r+1].reshape(28, 28), cmap='Greys', interpolation='nearest')
plt.show()
```

# MNIST Evaluation

```python
correct_prediction = tf.equal(tf.argmax(activation, 1), tf.argmax(y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, 'float'))
print('Accuracy : %s' % accuracy.eval(session=sess, feed_dict={x:mnist.test.images, y:mnist.test.labels}))
```
